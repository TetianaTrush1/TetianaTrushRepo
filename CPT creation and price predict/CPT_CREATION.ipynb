# Find last corect element in lists of Conditional probabilities with theorem of full probability:

```python
def FindLastelement(PriorPlst,CondPlst,priorProbabilityDisease):
    if len(PriorPlst) != len(CondPlst)+1 :
        return None
    else:
        cou = 0
        summa = 0
        while cou != len(CondPlst):
            summa += PriorPlst[cou]*CondPlst[cou]
            cou +=1
        last_element = (priorProbabilityDisease - summa)/PriorPlst[cou]
    return last_element
        
```

```python
p1n.append(FindLastelement(PP1,p1n,priorProbabilityDisease))
p2n.append(FindLastelement(PP2,p2n,priorProbabilityDisease))
p3n.append(FindLastelement(PP3,p3n,priorProbabilityDisease))
p4n.append(FindLastelement(PP4,p4n,priorProbabilityDisease))
p5n.append(FindLastelement(PP5,p5n,priorProbabilityDisease))
p6n.append(FindLastelement(PP6,p6n,priorProbabilityDisease))
p7n.append(FindLastelement(PP7,p7n,priorProbabilityDisease))
```

# Check full probability rule:


```python
def checkFullProbRule(PriorPlst,CondPlst,priorProbabilityDisease):
    if len(PriorPlst) != len(CondPlst) :
        return None
    else:
        cou = 0
        summa = 0
        while cou != len(PriorPlst):
            summa += PriorPlst[cou]*CondPlst[cou]
            cou +=1
        return summa == priorProbabilityDisease
```

```python
print(checkFullProbRule(PP1,p1n,priorProbabilityDisease))
print(checkFullProbRule(PP2,p2n,priorProbabilityDisease))
print(checkFullProbRule(PP3,p3n,priorProbabilityDisease))
print(checkFullProbRule(PP4,p4n,priorProbabilityDisease))
print(checkFullProbRule(PP5,p5n,priorProbabilityDisease))
print(checkFullProbRule(PP6,p6n,priorProbabilityDisease))
print(checkFullProbRule(PP7,p7n,priorProbabilityDisease))

```
